```python
from flair.data import Sentence
from flair.models import SequenceTagger

# load tagger
tagger = SequenceTagger.load("flair/pos-english")

# make example sentence
sentence = Sentence("I love Berlin.")

# predict NER tags
tagger.predict(sentence)

# print sentence
print(sentence)

# print predicted NER spans
print('The following NER tags are found:')
# iterate over entities and print
for entity in sentence.get_spans('pos'):
    print(entity)
```

```python
import nltk
from nltk.corpus import stopwords
stopwords = stopwords.words('english')
```

```python
verb_tags = ['VB', 'VBD', 'VBG', 'VBN', 'VBZ']
remove_tags = ['VBP']
```

```python
relations = []
flag = 1
len_ = len(entity_pos)
for idx in range(len_):
    if idx == (len_ - 1):
        break
    else:
        ###
        if flag == 1:
            st = entity_pos[idx]['pos']
            st_entity = entity_pos[idx]['entity']
            flag = 0
        ed = entity_pos[idx + 1]['pos']
        ed_entity = entity_pos[idx + 1]['entity']
        ###
        chunk = (record['title']+' '+record['summary'])[st: ed] + entity_pos[idx+1]['entity']
        sentence = Sentence(chunk)
        tagger.predict(sentence)
        tags = sentence.get_spans('pos')
        
        
        _ = []
        __ = []

        for tag in tags:
            if tag.tag in verb_tags:
                if tag.text not in stopwords:
                    __.append(tag.text)
        
        if __:
            _.append({entity_pos[idx]['ner_type']: entity_pos[idx]['entity']})
            _.append({'relations': __})
            _.append({entity_pos[idx + 1]['ner_type']: entity_pos[idx + 1]['entity']})
            relations.append(_)  
        
        ###
        st = ed
        st_entity = ed_entity
        
relations
```
